```python
# Impute missing values with mean for specific columns
columns_to_impute = ['health_score', 'freedom_of_information_score', 'education_expenditures_score',
                     'natural_resource_protection_score', 'land_rights_and_access_score', 'trade_policy_score',
                     'inflation_score', 'fiscal_policy_score', 'access_to_credit_score', 'gender_in_the_economy_score',
                     'girls_education_enrollment_rates_score']

for col in columns_to_impute:
    df[col].fillna(df[col].mean(), inplace=True)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Calculate correlation matrix (with numeric_only=True)
corr_matrix = df.corr(numeric_only=True)

# Create heatmap with annotations
plt.imshow(corr_matrix, cmap='coolwarm')
plt.colorbar()

# Add annotations manually
for i in range(len(corr_matrix.columns)):
    for j in range(len(corr_matrix.columns)):
        text = f"{corr_matrix.iloc[i, j]:.2f}"
        plt.annotate(text, xy=(j, i), ha='center', va='center')

plt.xticks(range(len(corr_matrix.columns)), corr_matrix.columns, rotation=45, ha='left')
plt.yticks(range(len(corr_matrix.columns)), corr_matrix.columns)
plt.title("Correlation Matrix for MCC Dataset")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Assuming your dataset is in a CSV file named "data.csv"
df = pd.read_csv("/content/mcc_2024_scorecard.csv")
```

```python
health_scores = df["health_score"]
government_effectiveness_scores = df["government_effectiveness_score"]
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(government_effectiveness_scores, health_scores)
plt.xlabel("Government Effectiveness Score")
plt.ylabel("Health Score")
plt.title("Correlation between Health Score and Government Effectiveness")
plt.show()
```

```python
# Calculate average health scores per country
avg_health_scores_per_country = df.groupby("country_name")["health_score"].mean()

# Create the bar plot
plt.figure(figsize=(12, 6))
plt.bar(avg_health_scores_per_country.index, avg_health_scores_per_country.values)
plt.xlabel("Country Name")
plt.ylabel("Average Health Score")
plt.title("Bar Plot of Average Health Scores per Country")
plt.xticks(rotation=90)  # Rotate x-axis labels for readability
plt.show()
```
